### Separar publicas e privadas
Observação: unidade de funcionamento não contém atributo direto para separar (caso necessário fazer relacionamento com outras tabelas).

```python
# Importar arquivos

import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

_2019_path = '2019/DADOS/'

sup_ies_path = 'SUP_IES_2019.CSV'
sup_curso_path = 'SUP_CURSO_2019.CSV'
sup_docente_path = 'SUP_DOCENTE_2019.CSV'
sup_aluno_path = 'SUP_ALUNO_2019.CSV'
```

```python
# Ler arquivos

encoding='ISO-8859-1'

sup_ies = pd.read_csv(_2019_path + sup_ies_path, '|', engine='python', encoding=encoding)
sup_curso = pd.read_csv(_2019_path + sup_curso_path, '|', engine='python', encoding=encoding)
sup_docente = pd.read_csv(_2019_path + sup_docente_path, '|', engine='python', encoding=encoding)
sup_aluno = pd.read_csv(_2019_path + sup_docente_path, '|', engine='python', encoding=encoding)

print('OK.')
```

```python
# Separar publico e privado

def split_public_private(df):
    public  = df[df["TP_CATEGORIA_ADMINISTRATIVA"].isin([1,2,3])]
    private = df[df["TP_CATEGORIA_ADMINISTRATIVA"].isin([4,5,6,7,8,9])]
    return public, private

sup_ies_public, sup_ies_private = split_public_private(sup_ies)
sup_curso_public, sup_curso_private = split_public_private(sup_curso)
sup_docente_public, sup_docente_private = split_public_private(sup_docente)
sup_aluno_public, sup_aluno_private = split_public_private(sup_aluno)

print('OK.')
```

```python
# Salvar

_2019_splitted_path = 'splitted/2019/dados/'
def save_splitted_to_csv(df_public, df_private, input_path):
    df_public.to_csv(_2019_splitted_path + input_path[0:-4] + "_PUBLICAS" + '.CSV', sep='|', encoding='utf-8')
    df_private.to_csv(_2019_splitted_path + input_path[0:-4] + "_PRIVADAS" + '.CSV', sep='|', encoding='utf-8')

save_splitted_to_csv(sup_ies_public, sup_ies_private, sup_ies_path)
save_splitted_to_csv(sup_curso_public, sup_curso_private, sup_curso_path)
save_splitted_to_csv(sup_docente_public, sup_docente_private, sup_docente_path)
save_splitted_to_csv(sup_aluno_public, sup_aluno_private, sup_aluno_path)

print('OK.')
```
