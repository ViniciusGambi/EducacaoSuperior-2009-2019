```python
import pandas as pd
#import numpy as np
#import matplotlib.pyplot as plt
#import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
file_base_name = 'SUP_ALUNO'
years = ['2009', '2019']

columns = {}
columns['2009'] = ['DS_CATEGORIA_ADMINISTRATIVA','NO_CURSO','CO_MODALIDADE_ENSINO','CO_GRAU_ACADEMICO','CO_ALUNO_SITUACAO','IN_ING_PROCESSO_SELETIVO','IN_ING_VESTIBULAR','IN_ING_ENEM','IN_ING_OUTRA_FORMA_SELECAO','IN_ING_PROCESSO_OUTRAS_FORMAS','IN_ALUNO_DEFICIENCIA','IN_RESERVA_VAGAS','IN_RESERVA_ENSINO_PUBLICO','IN_RESERVA_ETNICO','IN_RESERVA_DEFICIENCIA','IN_RESERVA_RENDA_FAMILIAR','IN_RESERVA_OUTROS','IN_FIES','IN_PROUNI_INTEGRAL','IN_PROUNI_PARCIAL','IN_SEXO_ALUNO','CO_COR_RACA_ALUNO','IN_CONCLUINTE']
columns['2019'] = ['NU_ANO_CENSO', 'TP_CATEGORIA_ADMINISTRATIVA', 'CO_CURSO', 'TP_MODALIDADE_ENSINO', 'TP_GRAU_ACADEMICO', 'TP_SITUACAO', 'IN_INGRESSO_PROCESSO_SELETIVO', 'IN_INGRESSO_VESTIBULAR', 'IN_INGRESSO_ENEM', 'IN_INGRESSO_OUTRA_FORMA', 'IN_DEFICIENCIA', 'IN_RESERVA_VAGAS', 'IN_RESERVA_ENSINO_PUBLICO', 'IN_RESERVA_ETNICO', 'IN_RESERVA_DEFICIENCIA', 'IN_RESERVA_RENDA_FAMILIAR', 'IN_RESERVA_OUTRA', 'IN_FIN_REEMB_FIES', 'IN_FIN_NAOREEMB_PROUNI_INTEGR', 'IN_FIN_NAOREEMB_PROUNI_PARCIAL', 'TP_SEXO', 'TP_COR_RACA', 'IN_CONCLUINTE']

encoding='ISO-8859-1'

students = {}
chunksize = 10**5
for year in years:
    students[year] = pd.DataFrame()
    for chunk in pd.read_csv('{}/dados/{}.CSV'.format(year, file_base_name), chunksize=chunksize, encoding=encoding, sep='|'):
        processed = chunk[columns[year]]
        students[year] = pd.concat([students[year], processed])
    
```

```python
new_columns = {
    'DS_CATEGORIA_ADMINISTRATIVA': 'TP_CATEGORIA_ADMINISTRATIVA',
    'CO_CURSO': 'CO_CURSO',
    'CO_MODALIDADE_ENSINO': 'TP_MODALIDADE_ENSINO',
    'CO_GRAU_ACADEMICO': 'TP_GRAU_ACADEMICO',
    'CO_ALUNO_SITUACAO': 'TP_SITUACAO',
    'IN_ING_PROCESSO_SELETIVO': 'IN_INGRESSO_PROCESSO_SELETIVO',
    'IN_ING_VESTIBULAR': 'IN_INGRESSO_VESTIBULAR',
    'IN_ING_ENEM': 'IN_INGRESSO_ENEM',
    'IN_ING_OUTRA_FORMA_SELECAO': '',
    'IN_ING_OUTRAS_FORMAS': 'IN_INGRESSO_OUTRA_FORMA',
    'IN_ALUNO_DEFICIENCIA': 'IN_DEFICIENCIA',
    'IN_RESERVA_VAGAS': 'IN_RESERVA_VAGAS',
    'IN_RESERVA_ENSINO_PUBLICO': 'IN_RESERVA_ENSINO_PUBLICO',
    'IN_RESERVA_ETNICO': 'IN_RESERVA_ETNICO',
    'IN_RESERVA_DEFICIENCIA': 'IN_RESERVA_DEFICIENCIA',
    'IN_RESERVA_RENDA_FAMILIAR': 'IN_RESERVA_RENDA_FAMILIAR',
    'IN_RESERVA_OUTROS': 'IN_RESERVA_OUTRA',
    'IN_FIES': 'IN_FIN_REEMB_FIES',
    'IN_PROUNI_INTEGRAL': 'IN_FIN_NAOREEMB_PROUNI_INTEGR',
    'IN_PROUNI_PARCIAL': 'IN_FIN_NAOREEMB_PROUNI_PARCIAL',
    'IN_SEXO_ALUNO': 'TP_SEXO',
    'CO_COR_RACA_ALUNO': 'TP_COR_RACA',
    'IN_CONCLUINTE': 'IN_CONCLUINTE'
}

students['2009'].rename(columns=new_columns, inplace=True)
students['2009']['NU_ANO_CENSO'] = pd.Series([2009 for x in range(len(students['2009'].index))])

students['2009']['TP_CATEGORIA_ADMINISTRATIVA'] = students['2009']['TP_CATEGORIA_ADMINISTRATIVA'].replace({1: 'Federal', 2: 'Estadual', 3: 'Municipal', 4: 'Privada', 5: 'Privada', 6: 'Privada'})
students['2009']['TP_SEXO'] = students['2009']['TP_SEXO'].replace({0: 'Masculino', 1: 'Feminino'})
```
