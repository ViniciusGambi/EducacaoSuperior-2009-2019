### Setup inicial

```python
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

_2009_path = '2009/DADOS/'

dm_ies_path = 'DM_IES.CSV'
dm_curso_path = 'DM_CURSO.CSV'
dm_docente_path = 'DM_DOCENTE.CSV'
dm_aluno_path = 'DM_ALUNO.CSV'
dm_unidade_funcionamento_path = 'DM_UNIDADE_FUNCIONAMENTO.CSV'
dm_vagas_ead_path = 'DM_VAGAS_INSCRITOS_EAD.CSV'
```

```python
# Ler arquivos

encoding='ISO-8859-1'

dm_ies = pd.read_csv(filepath_or_buffer=_2009_path + dm_ies_path, sep='|', encoding=encoding)
dm_curso = pd.read_csv(filepath_or_buffer=_2009_path + dm_curso_path, sep='|', encoding='ISO-8859-1')
dm_docente = pd.read_csv(filepath_or_buffer=_2009_path + dm_docente_path, sep='|', encoding='ISO-8859-1')
dm_aluno = pd.read_csv(filepath_or_buffer=_2009_path + dm_docente_path, sep='|', encoding='ISO-8859-1')
dm_unidade_funcionamento = pd.read_csv(filepath_or_buffer=_2009_path + dm_unidade_funcionamento_path, sep='|', encoding='ISO-8859-1')
dm_vagas_ead = pd.read_csv(filepath_or_buffer=_2009_path + dm_vagas_ead_path, sep='|', encoding='ISO-8859-1')

print('OK.')
```
